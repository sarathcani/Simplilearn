# PROJECT 3 CNN

## Part 1 : Importing modules

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
import cv2
import os
import matplotlib.pyplot as plt
from random import shuffle
```

## Part 2 : Setting the hyperparameters

```python
IMAGE_SIZE = 128
FC_SIZE = 32
```

## Part 3 : Read image data set

```python
DATA_DIR = os.path.join(os.getcwd() , 'datasets' , 'p_3')

TRAIN_DIR = os.path.join(DATA_DIR , 'train')
TEST_DIR = os.path.join(DATA_DIR , 'test')
AUG_DIR = os.path.join(DATA_DIR , 'aug')

categories = ['cats' , 'dogs']
```

```python
training_data = []
testing_data = []

def create_training_data():
    for category in categories :
        path = os.path.join(TRAIN_DIR , category)
        category_index = categories.index(category)
        for image in os.listdir(path):
            image_array = cv2.imread(os.path.join(path , image) )
            resized_array = cv2.resize(image_array , (IMAGE_SIZE , IMAGE_SIZE ))
            training_data.append([resized_array , category_index])
def create_testing_data():
    for category in categories :
        path = os.path.join(TEST_DIR , category)
        category_index = categories.index(category)
        for image in os.listdir(path):
            image_array = cv2.imread(os.path.join(path , image) )
            resized_array = cv2.resize(image_array , (IMAGE_SIZE , IMAGE_SIZE ))
            testing_data.append([resized_array , category_index])

create_training_data()
create_testing_data()

shuffle(training_data)
shuffle(testing_data)
```

```python
x_train = []
y_train = []

for features , label in training_data :
    x_train.append(features)
    y_train.append(label)
    
x_test = []
y_test = []

for features , label in testing_data :
    x_test.append(features)
    y_test.append(label)
    
x_train = np.array(x_train)
x_test = np.array(x_test)
```

```python
x_train.shape
```

## Part 4 : Run TensorFlow model

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.BatchNormalization(input_shape=(IMAGE_SIZE , IMAGE_SIZE , 3 , )))
model.add(tf.keras.layers.Conv2D(32 , kernel_size=(5,5) , activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2) , strides=2))
model.add(tf.keras.layers.Conv2D(64 , kernel_size=(5,5) , activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2) , strides=2))

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(FC_SIZE , activation='relu'))
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Dense(2 , activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam' , loss='categorical_crossentropy' , metrics=['accuracy'])
```

```python
# Fitting without image augmentation
model.fit(x_train , to_categorical(y_train) , epochs=100 , validation_data=(x_test , to_categorical(y_test)))
```

```python
# Fitting without image augmentation
model.fit(x_train , to_categorical(y_train) , epochs=200 , validation_data=(x_test , to_categorical(y_test)))
```

```python
# Fitting without image augmentation
model.fit(x_train , to_categorical(y_train) , epochs=300 , validation_data=(x_test , to_categorical(y_test)))
```
